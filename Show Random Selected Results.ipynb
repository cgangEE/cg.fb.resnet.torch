```lua
require 'torch'
require 'paths'
require 'cudnn'
require 'cunn'
require 'image'
require 'socket'

model = torch.load('checkpointFlickr200_1/model_38.t7'):cuda()

for i=11,9,-1 do
    print(model:remove(i))
end

function Test(imgName)


    model:evaluate()

    meanstd = {
        mean = { 0.485, 0.456, 0.406 },
        std = { 0.229, 0.224, 0.225 },
    }

    t = require './datasets/transforms'

    transform = t.Compose{
        t.Scale(256),
        t.ColorNormalize(meanstd),
        t.CenterCrop(224),
    }


    img = image.load(imgName, 3, 'float')
    img = transform(img)
    

    output = model:forward(img:cuda()):squeeze()
    return output, img
end


function outputSize(l)
    print(l.output:size())
end

function show(o)
    o = o.output
    o = o:view(o:size(2), o:size(1), o:size(3), o:size(4))
    itorch.image(o)
end


function sumUp(o)
    local x = o[{{},1,{},{}}]
    local y = torch.zeros( x:size()):cuda()

    for i = 1,(o:size(2)) do
        x = o[{{},i,{},{}}] 
        y = y + x
    end
    y = nn.utils.recursiveType(y, 'torch.FloatTensor')
    return y
end


function showFeatureMap(imgName)
    output, img = Test(imgName)

    itorch.image(img)
    
    l = model:get(8)
    o = l.output
    y  = sumUp(o)

    y = image.scale(y, img:size(3), img:size(2), 'bilinear')

    itorch.image(y)
end

for line in io.lines('flickr/flickr10000_group1.txt') do
    if math.random(1000) <= 10 then
        print(line)
        showFeatureMap('flickr/images/' .. line)
    end
end
showFeatureMap('./2.jpg')
```
